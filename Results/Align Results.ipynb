```python
import pandas as pd
import numpy as np
```

```python
file1 = "HAR {EUR} BIC.csv"
file2 = "HAR {NE} BIC.csv"
#file3 = "file3.csv"
```

```python
# Read CSV files, specifying columns B (index 1) and I (index 8)
df1 = pd.read_csv(file1, usecols=[1, 8], names=['date', 'test_qlike'], skiprows=1)
df2 = pd.read_csv(file2, usecols=[1, 8], names=['date', 'test_qlike'], skiprows=1)
#df3 = pd.read_csv(file3, usecols=[1, 8], names=['date', 'test_qlike'], skiprows=1)

# Convert date columns to datetime with dayfirst=True for DD/MM/YYYY format
df1['date'] = pd.to_datetime(df1['date'], dayfirst=True)
df2['date'] = pd.to_datetime(df2['date'], dayfirst=True)
#df3['date'] = pd.to_datetime(df3['date'], dayfirst=True)

# Set date as index for each dataframe
df1.set_index('date', inplace=True)
df2.set_index('date', inplace=True)
#df3.set_index('date', inplace=True)

# Rename columns to identify source
df1.rename(columns={'test_qlike': 'test_qlike_1'}, inplace=True)
df2.rename(columns={'test_qlike': 'test_qlike_2'}, inplace=True)
#df3.rename(columns={'test_qlike': 'test_qlike_3'}, inplace=True)

# Merge the three dataframes on date index (inner join to keep only common dates)
merged_df = df1.join(df2, how='inner')#.join(df3, how='inner')

# Calculate the average test_qlike across the three files
merged_df['avg_test_qlike'] = merged_df[['test_qlike_1', 'test_qlike_2']].mean(axis=1)

#merged_df['avg_test_qlike'] = merged_df[['test_qlike_1', 'test_qlike_2', 'test_qlike_3']].mean(axis=1)

# Create output dataframe with date and average
output_df = merged_df[['avg_test_qlike']].reset_index()
output_df.columns = ['date', 'avg_test_qlike']
```

```python
print(f"File 1 date range: {df1.index.min()} to {df1.index.max()} ({len(df1)} rows)")
print(f"File 2 date range: {df2.index.min()} to {df2.index.max()} ({len(df2)} rows)")
#print(f"File 3 date range: {df3.index.min()} to {df3.index.max()} ({len(df3)} rows)")
print(f"\nCommon date range: {output_df['date'].min()} to {output_df['date'].max()}")
print(f"Number of common dates: {len(output_df)}")
print(f"\nFirst few rows of output:")
print(output_df.head(10))
```

```python
output_file = "averaged_test_qlike.csv"
output_df.to_csv(output_file, index=False)
print(f"\nOutput saved to: {output_file}")
```
